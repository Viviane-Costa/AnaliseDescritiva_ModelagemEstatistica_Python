# **Medidas de tendência central**

As medidas de tendência central são estatísticas que resumem um conjunto de dados com um único valor que representa o centro da distribuição.

**Código**

# **Média**
**Definição:** É a soma de todos os valores dividida pelo número de observações.

Média.webp

onde $𝑥_𝑖$ são os valores dos dados e $𝑛$ é o número total de observações.

```python
print(np.mean(dados['gasolina_comum_preco_revenda_avg']))
print(np.mean(dados['gasolina_aditivada_preco_revenda_avg']))
print(np.mean(dados['etanol_hidratado_preco_revenda_avg']))
print(np.mean(dados['oleo_diesel_preco_revenda_avg']))
print(np.mean(dados['gas_cozinha_glp_preco_revenda_avg']))
```

```python
# Arredondar os valores de cada cálculo para 2 casas decimais
print(round(np.mean(dados['gasolina_comum_preco_revenda_avg']), 2))
print(round(np.mean(dados['gasolina_aditivada_preco_revenda_avg']), 2))
print(round(np.mean(dados['etanol_hidratado_preco_revenda_avg']), 2))
print(round(np.mean(dados['oleo_diesel_preco_revenda_avg']), 2))
print(round(np.mean(dados['gas_cozinha_glp_preco_revenda_avg']), 2))
```

# **Mediana**

A média é sensível a valores extremos (*outliers*) e pode muitas vezes não representar bem a tendência central da variável. Para solucionar essa limitação tem-se a Mediana, e como o próprio nome já entrega, representa o meio dos dados ordenados, indicando de fato a centralidade sem efeitos extremos.

mediana.webp

```python
print(round(np.median(dados['gasolina_comum_preco_revenda_avg']), 2))
print(round(np.median(dados['etanol_hidratado_preco_revenda_avg']), 2))
print(round(np.median(dados['oleo_diesel_preco_revenda_avg']), 2))
print(round(np.median(dados['gas_cozinha_glp_preco_revenda_avg']), 2))
```

# **Moda**

É a medida de tendência central que representa o valor mais frequente em um conjunto de dados. Em outras palavras, é o valor que ocorre com maior frequência.

```python
st.mode(dados['etanol_hidratado_preco_revenda_avg'])
```

# **Medidas de Dispersão**

# **Variância**
É uma medida de dispersão que indica o quanto os valores de um conjunto de dados se afastam da média. Em outras palavras, ela mede a variabilidade dos dados.

Variancia.webp

```python
print(round(st.variance(dados['gasolina_comum_preco_revenda_avg']),2))
print(round(st.variance(dados['etanol_hidratado_preco_revenda_avg']),2))
```

# **Desvio padrão**

Dado que o cálculo da variância eleva-se ao quadrado, para obter uma real comparação da variação dos dados em relação a média, aplica-se a raiz quadrada na variância, chegando assim no desvio padrão.

Desvio.webp

```python
print(round(st.sqrt(st.variance(dados['gasolina_comum_preco_revenda_avg'])),2))
print(round(st.sqrt(st.variance(dados['etanol_hidratado_preco_revenda_avg'])),2))
```

# **Coeficiente de Variação**

O coeficiente de variação (CV) é uma medida relativa de dispersão que expressa a variabilidade de um conjunto de dados em relação à sua média. Ele é útil para comparar a variabilidade de diferentes conjuntos de dados, especialmente quando as unidades de medida são diferentes.

cof de variacao.webp

```python
print(round(st.sqrt(st.variance(dados['gasolina_comum_preco_revenda_avg']))/st.mean(dados['gasolina_comum_preco_revenda_avg']),2))
print(round(st.sqrt(st.variance(dados['gasolina_comum_preco_revenda_avg']))/st.mean(dados['gasolina_comum_preco_revenda_avg']),2))
```

# **Customização e Estilo**
Você pode personalizar ainda mais seus gráficos para torná-los mais atraentes.

**Seaborn Customizado**
Seaborn permite ajustes de estilo, como adicionar uma grid de fundo, mudar o tema e ajustar as cores.

**Plotly Customizado**

No Plotly, você pode adicionar interatividade, títulos dinâmicos, e até mesmo legendas que mudam conforme o gráfico é manipulado.

```python
fig = px.scatter(dados, x='etanol_hidratado_preco_revenda_avg',
                 y='gasolina_comum_preco_revenda_avg',
                 color='ano',
                 hover_data=['ano', 'etanol_hidratado_preco_revenda_avg'],
                 title='Preço de Gasolina vs Preço de Etanol')

# Customizando o layout
fig.update_layout(
    xaxis_title='Preço do Etanol',
    yaxis_title='Preço da Gasolina',
    template='plotly_dark'
)

fig.show()
```

# **Análise de Regressão**

# **Regressão Linear Simples**

Se você deseja examinar a relação entre duas variáveis, por exemplo, o preço de um tipo de combustível (como a gasolina comum) em função de outro (como o preço do etanol), pode usar a regressão linear simples.

O conjunto de dados habitacionais de Boston.

O Boston Housing Dataset é derivado de informações coletadas pelo Serviço de Censo dos EUA sobre moradias na área de Boston MA. O seguinte descreve as colunas do conjunto de dados:

Features

RM: número médio de quartos por habitação

LSTAT: porcentagem da população considerada de status inferior

PTRATIO: relação aluno-professor por município

Target

MEDV: valor médio de casas ocupadas pelos proprietários


```python
dados_2 = pd.read_csv('/content/housing.csv', sep=';')
```

```python
dados_2.head()
```

```python
# Selecionando apenas as colunas desejadas
dados_selecionados = dados_2[['RM', 'LSTAT', 'MEDV']]

# Exibindo as primeiras linhas do novo DataFrame
print(dados_selecionados.head())
```

```python
sns.pairplot(dados_selecionados)
```

```python
# Variáveis para a regressão linear simples
X_simples = dados_2['RM']  # Variável preditora
y = dados_2['MEDV']        # Variável resposta (target)

# Adicionar uma constante para o modelo
X_simples = sm.add_constant(X_simples)

# Ajustar o modelo de regressão
modelo_simples = sm.OLS(y, X_simples).fit()

# Resumo dos resultados
print(modelo_simples.summary())
```

# **Regressão Linear Múltipla**

Se você quer analisar o impacto de várias variáveis no preço de um combustível, você pode realizar uma regressão linear múltipla.

```python
# Variáveis para a regressão linear múltipla
X_multipla = dados_2[['RM', 'LSTAT']]  # Variáveis preditoras
y = dados_2['MEDV']                               # Variável resposta (target)

# Adicionar uma constante para o modelo
X_multipla = sm.add_constant(X_multipla)
# Ajustar o modelo de regressão
modelo_multipla = sm.OLS(y, X_multipla).fit()

# Resumo dos resultados
print(modelo_multipla.summary())
```
